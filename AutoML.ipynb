```python
from sklearn import datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = datasets.load_iris()
```

```python
iris
```

```python
X=iris.data
y=iris.target
```

```python
# Convert the data (features) and target (labels) to a pandas DataFrame
iris_data = pd.DataFrame(data=iris['data'], columns=iris['feature_names'])
iris_data['target'] = iris['target']
```

```python
import h2o
from h2o.automl import H2OAutoML

# Start H2O cluster
h2o.init()
```

```python
# Convert the pandas DataFrame to an H2O Frame
iris_h2o = h2o.H2OFrame(iris_data)

# Check the data
iris_h2o.head()
```

```python
from h2o.automl import H2OAutoML

# Set the target column and split the data into features and target
target_column = 'target'
x = iris_h2o.columns[:-1]  # Features (all columns except the target)
y = target_column           # Target (the last column)

# Split the dataset into train and test sets
train, test = iris_h2o.split_frame(ratios=[.8], seed=1234)
```

```python
# Initialize H2O AutoML (you can limit max_models or time)
aml = H2OAutoML(max_runtime_secs=300, seed=1)

# Train the model on the training set
aml.train(x=x, y=y, training_frame=train)
```

```python
# Access the best model (leader)
best_model = aml.leader

# Explain the model
best_model.explain(test)
```
